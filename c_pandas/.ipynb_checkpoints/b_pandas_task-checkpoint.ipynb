```python
import pandas as pd

# 자료 파일 불러옴
h_df = pd.read_csv('./datasets/happiness_report_2023.csv')
display(h_df)
```

```python
# 자료 정보
h_df.info()

# 자료형
h_df.dtypes
```

```python
# 숫자형 컬럼에 대한 정보 계산(개수, 평균, 표준편차, 최솟값, 하위25%, 중앙값, 상위 25%, 최댓값)
h_df.describe().T
```

```python
import numpy as np
from scipy.stats import iqr

# 하위 25%
h_df_Q1 = np.percentile(h_df.happiness_score, 25)
# 상위 25%
h_df_Q3 = np.percentile(h_df.happiness_score, 75)
# print(h_df_Q1, h_df_Q3)

# iqr: 정상치 범위
# 1.
# iqr_value = h_df_Q3 - h_df_Q1
# print(iqr_value)
# 2. 
iqr_value = iqr(h_df.happiness_score)
print(iqr_value)

lower_bound = h_df_Q1 - 1.5 * iqr_value
upper_bound = h_df_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
# 정상치 범위 내의 정보만 출력
condition1 = h_df.happiness_score >= lower_bound
condition2 = h_df.happiness_score <= upper_bound
condition = condition1 & condition2

h_df_condition = h_df[condition]
h_df_condition
```

```python
# 누락된 데이터(NaN) 확인(.isna(): NaN을 True로 함)하여 그 개수를 합산(.sum())
h_df.isna().sum()
```

```python
# 중복값이 있는지 확인하여 개수를 합산
h_df.duplicated().sum()
```

```python
display(h_df[['gdp_per_capita']].describe().T)
# 1인당 국내총생산(GDP)에 따른 행복지수의 차이 -> 영향 있는 듯?

print(h_df[h_df['gdp_per_capita'] < 1.5]['happiness_score'].mean())
print(h_df[h_df['gdp_per_capita'] > 1.5]['happiness_score'].mean())
```

```python
display(h_df[['freedom_to_make_life_choices']].describe().T)
# 삶의 선택에 있어 자유로움의 척도에 따른 행복지수

print(h_df[h_df['freedom_to_make_life_choices'] < 0.55]['happiness_score'].mean())
print(h_df[h_df['freedom_to_make_life_choices'] > 0.55]['happiness_score'].mean())
```

```python
h_df['freedom_to_make_life_choices'].duplicated().sum()
```

```python
h_mean_df = h_df.groupby('freedom_to_make_life_choices').mean('happiness_score').reset_index()
h_mean_df
```

```python
# matplotlib 라이브러리의 pyplot 모듈을 불러옴
import matplotlib.pyplot as plt

# subplots: 여러 개의 그래프(서브플롯)를 한 Figure 안에 만들 때 사용
# 2, 2: 2행 2열 형태로 총 4개의 subplot 생성
# figsize=(12, 9): 전체 그림의 크기를 인치 단위로 지정(너비12, 높이9)
# fig: 전체 Figure(도화지)에 해당하는 객체
# axes: 각 subplot의 축(ax) 객체를 2차원 배열로 담고 있음
#       → axes[0,0], axes[0,1], axes[1,0], axes[1,1]로 접근 가능
fig, axes = plt.subplots(2, 2, figsize=(12, 9))

# h_mean_df.plot(...): pandas의 DataFrame plotting 기능 사용
# title: 제목 / x=...: x축에 사용할 컬럼명 / y=...: y축에 사용할 컬럼명
# kind='line': 선형(line) 그래프 형태로 표시 / gird=True: 그래프 배경에 격자(grid)표시
# ax=axes[0, 0]: 생성한 4개의 subplot 중 왼쪽 위(0행 0열)에 그리도록 지정
h_mean_df.plot(title='mean', x='freedom_to_make_life_choices', y='happiness_score', fontsize='7', kind='line', grid=True, ax=axes[0, 0])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

bins = 10
cats = pd.cut(h_df['freedom_to_make_life_choices'], bins=bins)
h_mean_df = (h_df.groupby(cats)['happiness_score']
            .mean()
            .reset_index()
            .rename(columns={'freedom_to_make_life_choices': 'freedom_bin',
                            'happiness_score': 'happiness_mean'}))

ax = h_mean_df.plot(x='freedom_bin', y='happiness_mean',
                    kind='line', marker='o', grid=True, rot=45,
                    title='Mean Happiness by Freedom (binned)')
# plt.tight_layout()
plt.show()
```
